```python
import plotly.graph_objects as go 
import numpy  as np
```

```python
normal_distribution = np.random.normal(-10, 10, 10000)
```

```python
fig = go.Figure(
    data=[
        go.Scatter(
            x=list(range(len(normal_distribution))),
            y=normal_distribution,
            mode='lines+markers'  # both lines and points
        )
    ]
)
fig.show()
```

```python
# Your existing histogram
fig = go.Figure(
    data=[
        go.Histogram(
            x=normal_distribution,
            histnorm='probability density',  # Add this to match the curve scale
            opacity=0.7
        )
    ]
)

# Add the bell curve line on top
mean = -10
std_dev = 10
x_curve = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y_curve = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_curve - mean) / std_dev)**2)

fig.add_trace(go.Scatter(
    x=x_curve,
    y=y_curve,
    mode='lines',
    name='Bell Curve',
    line=dict(color='red', width=3)
))

fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Parameters from your distribution
mean = -10
std_dev = 10

# Create x values for the curve (range around the mean)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the normal distribution curve (probability density function)
y = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

# Create the plot
fig = go.Figure(
    data=[
        go.Scatter(
            x=x,
            y=y,
            mode='lines',
            name=f'Normal Distribution (μ={mean}, σ={std_dev})'
        )
    ]
)

fig.update_layout(
    title="Normal Distribution Bell Curve",
    xaxis_title="Value",
    yaxis_title="Probability Density"
)

fig.show()
```

```python
from scipy import stats

x_values = np.linspace(-50, 30, 100)
density = stats.norm.pdf(x_values, loc=-10, scale=10)  # theoretical density

fig = go.Figure(
    data=[
        go.Scatter(
            x=x_values,
            y=density,
            mode='lines'
        )
    ]
)
fig.show()
```

```python
# Method 1: Create bins and count frequencies manually
bins = np.linspace(-50, 30, 50)  # Create bins
counts, bin_edges = np.histogram(normal_distribution, bins=bins)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2  # Get bin centers

fig = go.Figure(
    data=[
        go.Scatter(
            x=bin_centers,  # X: the value ranges
            y=counts,       # Y: frequency of each range
            mode='lines+markers'
        )
    ]
)
fig.show()
```
